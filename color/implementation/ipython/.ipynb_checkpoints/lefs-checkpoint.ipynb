# Luminous Efficiency Functions

The [luminous efficiency function](http://en.wikipedia.org/wiki/Luminosity_function) characterizes the average spectral sensitivity of human visual perception of brightness. [Brightness](http://en.wikipedia.org/wiki/Brightness) is defined as the attribute of a visual sensation according to which an area appears to emit more or less light. [Photometry](http://goo.gl/JNMw6u) is the science of measurement of the intensity of light in terms of its perceived brightness to the human visual system.

## Photometry & Radiometry

Photometry is analogous to [radiometry](http://en.wikipedia.org/wiki/Radiometry) but is dedicated to the measurement of electromatic radiation frequencies over the visible spectrum. Photometric quantities are weighted accordingly to human visual system spectral sensitivity within the wavelength range 360-780 nanometres (nm) while radiometric quantities represent unweighted absolute power within the wavelength range 0.01-1000 micrometres (mm).

## Photoreceptors

The human eye has two different types of [photoreceptor cells](http://en.wikipedia.org/wiki/Photoreceptor_cell):

* The [cone cells](http://en.wikipedia.org/wiki/Cone_cell) are mediating the [photopic vision](http://en.wikipedia.org/wiki/Photopic_vision), responsible for color perception. In the photoreceptor layer of the [retina](http://en.wikipedia.org/wiki/Retina), around $6.8\cdot 10^6$ cone cells of types $S$, $M$, $L$ are measuring light with respective peak absorption at wavelengths of about 420 nm, 534 nm and 564 nm. Photopic vision luminance levels are defined for $L_v>10\ cd\cdot m^{-2}$.
* The [rod cells](http://en.wikipedia.org/wiki/Rod_cell) are mediating the [scotopic vision](http://en.wikipedia.org/wiki/Scotopic), the vision under dim light conditions. There are an average of $110-125\cdot 10^6$ rod cells in the retina although none are located in the central [fovea](http://en.wikipedia.org/wiki/Fovea) region. Scotopic vision luminance levels are defined for $L_v<0.01\ cd\cdot m^{-2}$.

## Luminous Efficiency Functions

[Color](https://github.com/KelSolaar/Color/) provides the following photopic and scotopic luminous efficiency functions:

```python
import color

print(sorted(color.PHOTOPIC_LEFS.keys()))
print(sorted(color.SCOTOPIC_LEFS.keys()))
```

### CIE 1924 Photopic Standard Observer

The *CIE 1924 Photopic Standard Observer* luminous efficiency function $V(\lambda)$ was established by the *CIE* in 1924 for a 2° angular subtense viewing field. It was derived from several independant experiments whose results were not averaged but weight assembled by *Gibson* and *Tyndall* (1923) from very different sets of data <a name="back_reference_2"><a href="#reference_2">[2]</a></a>:

* Minimum flicker experiments by *Coblentz* and *Emerson* (1918)
* Step-by-step experiments by *Hyde*, *Forsythe* and *Cady* (1918) and *Gibson* and *Tyndall* (1923)
* Direct brightness matching experiments by *Hyde*, *Forsythe* and *Cady* (1918) and *Hartman* (1918)

The angular subtense viewing field size varied among those experiments and the surround ﬁeld was not always of constant luminance, if present at all. The assembly of data gathered from very different experiments has subsequently affected the effectiveness of the luminous efficiency function $V(\lambda)$ <a name="back_reference_3"><a href="#reference_3">[3]</a></a>
.

Photopic luminous flux $P_v$ and radiant flux $P_e$ are related by the following equation <a name="back_reference_2"><a href="#reference_2">[2]</a></a>
:

$$
P_v=K_m\int_\lambda P_{e,\lambda}V(\lambda)d\lambda
$$

Accordingly, the photopic luminous efficacy $K$ defined as follows:

$$
K(\lambda)=\cfrac{P_{v,\lambda}}{P_{e,\lambda}}
$$

is equal to:

$$
K(\lambda)=K_mV(\lambda)
$$

with $K_m=683\ lm\cdot W^-1$ at $\lambda_d=555\ nm$.

The *CIE 1924 Photopic Standard Observer* luminous efficiency function $V(\lambda)$ has the following range and increments:

```python
color.PHOTOPIC_LEFS["CIE 1924 Photopic Standard Observer"].shape
```

```python
from color.implementations.matplotlib.plots import *

multi_spd_plot(
    [color.PHOTOPIC_LEFS["CIE 1924 Photopic Standard Observer"]],
    y_label="Luminous Efficiency",
    legend_location="upper right",
    y_tighten=True,
    margins=[0., 0., 0., .1])
```

### CIE 1951 Scotopic Standard Observer

The *CIE 1951 Scotopic Standard Observer* luminous efficiency function $V^\prime(\lambda)$ was established by the *CIE* in 1951 and is a relative function of wavelength with peak at $\lambda_d=507\ nm$.

Scotopic luminous flux $P_v^\prime$ and radiant flux $P_e$ are related by the following equation <a name="back_reference_2"><a href="#reference_2">[2]</a></a>:

$$
P_v^\prime=K_m\int_\lambda P_{e,\lambda}^\prime V(\lambda)d\lambda
$$

Accordingly, the scotopic luminous efficacy $K^\prime$ defined as follows:

$$
K^\prime(\lambda)=\cfrac{P_{v,\lambda}^\prime}{P_{e,\lambda}}
$$

is equal to:

$$
K^\prime(\lambda)=K_m^\prime V(\lambda)
$$

with $K_m=1700\ lm\cdot W^-1$ at $\lambda_d=507\ nm$.

The *CIE 1951 Scotopic Standard Observer* luminous efficiency function $V^\prime(\lambda)$ has the following range and increments:

```python
color.SCOTOPIC_LEFS["CIE 1951 Scotopic Standard Observer"].shape
```

```python
multi_spd_plot(
    [color.SCOTOPIC_LEFS["CIE 1951 Scotopic Standard Observer"]],
    y_label="Luminous Efficiency",
    legend_location="upper right",
    y_tighten=True,
    margins=[0., 0., 0., .1])
```

### Mesopic Vision

[Mesopic vision](http://en.wikipedia.org/wiki/Mesopic) is a combination of photopic vision and scotopic vision and is defined for for $L_v\pm 0.2\ cd\cdot m^{-2}$. Although there is no standard for mesopic vision, a weighting function $V_m(\lambda)$ is defined as follows <a name="back_reference_4"><a href="#reference_4">[4]</a></a>:

$$
V_m(\lambda)=(1-x)V^\prime(\lambda)+xV(\lambda)
$$

where $x$ is function of photopic luminance $L_p$ and can have value of:

| $L_p$ | Blue-heavy<br>(MOVE) | Blue-heavy<br>(LRC) | Red-heavy<br>(MOVE) | Red-heavy<br>(LRC) |
|:-----:|:--------------------:|:-------------------:|:-------------------:|:------------------:|
| 0.01  | 0.13                 | 0.04                | 0.00                | 0.01               |
| 0.1   | 0.42                 | 0.28                | 0.34                | 0.11               |
| 1.0   | 0.70                 | 1.00                | 0.68                | 1.00               |
| 10    | 0.98                 | 1.00                | 0.98                | 1.00               |

The [MOVE](http://www.move.vu.nl/en/) is an european research consortium and [LRC](http://www.lrc.rpi.edu/) stands for Lighting Research Center.

The *color.mesopic_weighting_function* implements support for mesopic vision following the above equation:

```python
color.mesopic_weighting_function(wavelength=500, Lp=0.2)
```

We can use different parameters for $x$:

```python
color.mesopic_weighting_function(500, 0.2, source="Red Heavy", method="LRC")
```

Conveniently the *color.mesopic_luminous_efficiency_function* definition implements the creation of a mesopic luminous efficiency function based on *color.mesopic_weighting_function* definition:

```python
# Plotting a mesopic luminous efficiency function with photopic luminance of 0.2.
mesopic_luminous_efficiency_function = color.mesopic_luminous_efficiency_function(0.2)

multi_spd_plot(
    [mesopic_luminous_efficiency_function,
     color.PHOTOPIC_LEFS["CIE 1924 Photopic Standard Observer"],
     color.SCOTOPIC_LEFS["CIE 1951 Scotopic Standard Observer"]],
    y_label="Luminous Efficiency",
    legend_location="upper right",
    y_tighten=True,
    margins=[0., 0., 0., .1])
```

### Judd Modified CIE 1951 Photopic Standard Observer

In 1951, *Judd* proposed a modified *CIE 1924 Photopic Standard Observer* luminous efficiency function $V(\lambda)$ because of effectiveness issues in the blue end of the visible spectrum leading to inadequacies in the [*CIE 1931 2° Standard Observer*](cmfs.ipynb#CIE-1931-2°-Standard-Observer) $\bar{x}(\lambda)$,$\bar{y}(\lambda)$,$\bar{z}(\lambda)$ color matching functions <a name="back_reference_5"><a href="#reference_5">[5]</a></a>.

The *Judd Modified CIE 1951 Photopic Standard Observer* luminous efficiency function $V(\lambda)$ has the following range and increments:

```python
color.PHOTOPIC_LEFS["Judd Modified CIE 1951 Photopic Standard Observer"].shape
```

```python
multi_spd_plot(
    [color.PHOTOPIC_LEFS["Judd Modified CIE 1951 Photopic Standard Observer"]],
    y_label="Luminous Efficiency",
    legend_location="upper right",
    y_tighten=True,
    margins=[0., 0., 0., .1])
```

The modification *Judd* proposed is in the wavelength region below 460 nm, where the blue sensitivity was slightly increased:

```python
multi_spd_plot(
    [color.PHOTOPIC_LEFS["CIE 1924 Photopic Standard Observer"],
     color.PHOTOPIC_LEFS["Judd Modified CIE 1951 Photopic Standard Observer"]],
    y_label="Luminous Efficiency",
    legend_location="upper right",
    y_tighten=True,
    margins=[0., 0., 0., .1])
```

The *CIE 1964 Photopic 10° Standard Observer* luminous efficiency function $V_{10}(\lambda)$ has the following range and increments:
